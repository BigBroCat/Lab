```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.datasets import fetch_lfw_people
import matplotlib.pyplot as plt
from sklearn.utils.class_weight import compute_class_weight

np.random.seed(20) # Set the seed value of the random number generator to ensure that the sequence of random numbers generated is the same every time the code is run.

# load the LFW data (Check the loading function and provide appropriate parameters)
lfw_people = fetch_lfw_people(min_faces_per_person=70, resize=0.4)
X = lfw_people.data # Image data is assigned to X
X_plot = lfw_people.images # Face image data is assigned to X
y = lfw_people.target # The target label is assigned to the variable y
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, stratify=y, random_state=42) # Divide the data set into training set and test set
target_names = lfw_people.target_names # The target name is assigned to the variable target_names
num_classes = max(y)+1

# Define how many images you want to display
n_images = 6

def sigmoid(self, x):
    return 1 / (1 + np.exp(-x))

# Create a grid plot with matplotlib
fig, ax = plt.subplots(1, n_images, figsize=(15, 10))
for i in range(n_images):
    ax[i].imshow(X_plot[i], cmap='gray')  # Display images in grayscale
    ax[i].set_title(target_names[y[i]])  # Set the title to the person's name
    ax[i].axis('off')  # Turn off axis numbering
plt.show()

# Define the Restricted Boltzmann Machine
class RestrictedBoltzmannMachine:
    def __init__(self, num_visible, num_hidden):
        # First, initialize the weights and biases.
        # Typically, weights are initialized to small random values, and biases can be initialized to zero.
        self.num_visible = num_visible
        self.num_hidden = num_hidden
        self.weights = np.random.normal(0, 0.1, size=(num_visible, num_hidden))
        self.visible_bias = np.zeros(num_visible) # ceate an array consists of '0', its size is num_visable
        self.hidden_bias = np.zeros(num_hidden)


    # Define the sigmoid function
    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x)) # Map the input value to the interval (0, 1)

    # Implement a step of the Contrastive Divergence (CD) algorithm to update the weights and biases.
    def contrastive_divergence(self, batch, learning_rate, k=1):

        positive_hidden_probabilities = self.sigmoid(np.dot(batch, self.weights) + self.hidden_bias)

        positive_hidden_states = (
                    positive_hidden_probabilities > np.random.rand(*positive_hidden_probabilities.shape)).astype(float)

        negative_visible_probabilities = self.sigmoid(
            np.dot(positive_hidden_states, self.weights.T) + self.visible_bias)

        negative_hidden_probabilities = self.sigmoid(
            np.dot(negative_visible_probabilities, self.weights) + self.hidden_bias)

        weight_update = np.dot(batch.T, positive_hidden_probabilities) - np.dot(negative_visible_probabilities.T,
                                                                                negative_hidden_probabilities)
        self.weights += learning_rate * weight_update / batch.shape[0]

        self.visible_bias += learning_rate * np.mean(batch - negative_visible_probabilities, axis=0)
        self.hidden_bias += learning_rate * np.mean(positive_hidden_probabilities - negative_hidden_probabilities,
                                                    axis=0)

        error = np.mean((batch - negative_visible_probabilities) ** 2)
        return error

    # Define training process. And also set some default value for the parameter.
    def train(self, data, learning_rate=0.01, num_epochs=1000, batch_size=32):
        for epoch in range(num_epochs):
            np.random.shuffle(data)
            total_batches = 0 # Record the total number of batches processed
            total_error = 0 # Cumulative reconstruction error for each batch
            for i in range(0, data.shape[0], batch_size):
                batch = data[i:i + batch_size]
                total_batches += 1
                total_error += self.contrastive_divergence(batch, learning_rate)

            avg_error = total_error / total_batches
            if (epoch + 1) % 10 == 0:
                print(f"Epoch {epoch + 1}, Average Reconstruction Error: {avg_error}")

                # show the dynamic figure directly
                if epoch + 1 in [100, 500, 1000]:
                    sample_reconstructed = self.reconstruct(data[:5])
                    print(sample_reconstructed)

                    plot_images(data[:5], sample_reconstructed, num_images=5)

    # Define the process calculating the probabilities for hidden nodes.
    def extract(self, data):
        hidden_probabilities = self.sigmoid(np.dot(data, self.weights) + self.hidden_bias)
        return hidden_probabilities

    def reconstruct(self, data):
        hidden_probabilities = self.extract(data)
        hidden_states = (hidden_probabilities > np.random.rand(*hidden_probabilities.shape)).astype(float)
        reconstructed_data = self.sigmoid(np.dot(hidden_states, self.weights.T) + self.visible_bias)
        return reconstructed_data


# Define the plot function for reconstruction.
def plot_images(original, reconstructed, num_images): # function definition
    plt.figure(figsize=(10, 2 * num_images)) # Create a graphical interface
    for i in range(num_images): # Loop through images
        # Show original image
        plt.subplot(num_images, 2, 2*i + 1)
        plt.imshow(original[i].reshape(50, 37), cmap='gray')
        plt.title("Original")
        plt.axis('off')
        # Show reconstructed image
        plt.subplot(num_images, 2, 2*i + 2)
        plt.imshow(reconstructed[i].reshape(50, 37), cmap='gray')
        plt.title("Reconstructed")
        plt.axis('off')
    # Adjust layout and display graphics
    plt.tight_layout()
    plt.show()

# Initialize and train the RBM model
num_visible = X_train.shape[1]
num_hidden = 128  # The parameters that can be adjusted depend on the complexity of the model and the characteristics of the training data.
rbm = RestrictedBoltzmannMachine(num_visible, num_hidden) # Create an instance of the RestrictedBoltzmannMachine class.

# Call the train method of the rbm instance to train the model
rbm.train(X_train, learning_rate=0.01, num_epochs=5000, batch_size=32)

# Plot the sampe reconstructed by the trained model
num_plot = 5
test_data = X_test[:num_plot]  # Plot the sampe reconstructed by the trained model
reconstructed_data = rbm.reconstruct(test_data)

plot_images(test_data, reconstructed_data, num_images=num_plot)
```

```python
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt


# Define MLP classifier
class MLPClassifier:
    def __init__(self, input_dim, hidden_dim, output_dim, learning_rate=0.01):
        self.input_dim = input_dim
        self.hidden_dim = hidden_dim
        self.output_dim = output_dim
        self.learning_rate = learning_rate

        # Initialize weights and biases
        self.weights1 = np.random.normal(0, 0.1, (input_dim, hidden_dim))
        self.bias1 = np.zeros(hidden_dim)
        self.weights2 = np.random.normal(0, 0.1, (hidden_dim, output_dim))
        self.bias2 = np.zeros(output_dim)

    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))

    def softmax(self, x):
        exp_x = np.exp(x - np.max(x, axis=1, keepdims=True))
        return exp_x / np.sum(exp_x, axis=1, keepdims=True)

    def forward(self, x):
        self.z1 = np.dot(x, self.weights1) + self.bias1
        self.a1 = self.sigmoid(self.z1)
        self.z2 = np.dot(self.a1, self.weights2) + self.bias2
        self.a2 = self.softmax(self.z2)
        return self.a2

    def backward(self, x, y, output):
        m = x.shape[0]
        y_one_hot = np.eye(self.output_dim)[y]

        dz2 = output - y_one_hot
        dw2 = np.dot(self.a1.T, dz2) / m
        db2 = np.sum(dz2, axis=0) / m

        dz1 = np.dot(dz2, self.weights2.T) * self.a1 * (1 - self.a1)
        dw1 = np.dot(x.T, dz1) / m
        db1 = np.sum(dz1, axis=0) / m

        # Update weights and biases
        self.weights2 -= self.learning_rate * dw2
        self.bias2 -= self.learning_rate * db2
        self.weights1 -= self.learning_rate * dw1
        self.bias1 -= self.learning_rate * db1

    def train(self, x, y, epochs=1000):
        for epoch in range(epochs):
            output = self.forward(x)
            self.backward(x, y, output)
            if (epoch + 1) % 100 == 0:
                loss = -np.mean(np.log(output[range(y.shape[0]), y]))
                print(f"Epoch {epoch + 1}, Loss: {loss}")

    def predict(self, x):
        output = self.forward(x)
        return np.argmax(output, axis=1)


# Data preprocessing and loading
lfw_data = datasets.fetch_lfw_people(min_faces_per_person=70, resize=0.4)
features = lfw_data.data
labels = lfw_data.target

# Normalized data
scaler = StandardScaler()
features = scaler.fit_transform(features)

# Split the dataset
X_train, X_val, y_train, y_val = train_test_split(features, labels, test_size=0.2, random_state=42)

# Train the MLP classifier
mlp = MLPClassifier(input_dim=X_train.shape[1], hidden_dim=128, output_dim=len(np.unique(labels)), learning_rate=0.01)
mlp.train(X_train, y_train, epochs=2000)

# Evaluate MLP model
mlp_predictions = mlp.predict(X_val)
print("MLP Model Performance:")
accuracy = accuracy_score(y_val, mlp_predictions)
precision = precision_score(y_val, mlp_predictions, average='weighted')
recall = recall_score(y_val, mlp_predictions, average='weighted')
f1 = f1_score(y_val, mlp_predictions, average='weighted')
print(f"Accuracy: {accuracy}, Precision: {precision}, Recall: {recall}, F1 Score: {f1}")

# Train SVM classifier
svm = SVC(kernel='linear', C=1.0, random_state=42)
svm.fit(X_train, y_train)
svm_predictions = svm.predict(X_val)

# Evaluate SVM model
print("SVM Model Performance:")
accuracy = accuracy_score(y_val, svm_predictions)
precision = precision_score(y_val, svm_predictions, average='weighted')
recall = recall_score(y_val, svm_predictions, average='weighted')
f1 = f1_score(y_val, svm_predictions, average='weighted')
print(f"Accuracy: {accuracy}, Precision: {precision}, Recall: {recall}, F1 Score: {f1}")
```

```python
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt


# Define MLP classifier
class MLPClassifier:
    def __init__(self, input_dim, hidden_dim, output_dim, learning_rate=0.01):
        self.input_dim = input_dim
        self.hidden_dim = hidden_dim
        self.output_dim = output_dim
        self.learning_rate = learning_rate

        # Initialize weights and biases
        self.weights1 = np.random.normal(0, 0.1, (input_dim, hidden_dim))
        self.bias1 = np.zeros(hidden_dim)
        self.weights2 = np.random.normal(0, 0.1, (hidden_dim, output_dim))
        self.bias2 = np.zeros(output_dim)

    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))

    def softmax(self, x):
        exp_x = np.exp(x - np.max(x, axis=1, keepdims=True))
        return exp_x / np.sum(exp_x, axis=1, keepdims=True)

    def forward(self, x):
        self.z1 = np.dot(x, self.weights1) + self.bias1
        self.a1 = self.sigmoid(self.z1)
        self.z2 = np.dot(self.a1, self.weights2) + self.bias2
        self.a2 = self.softmax(self.z2)
        return self.a2

    def backward(self, x, y, output):
        m = x.shape[0]
        y_one_hot = np.eye(self.output_dim)[y]

        dz2 = output - y_one_hot
        dw2 = np.dot(self.a1.T, dz2) / m
        db2 = np.sum(dz2, axis=0) / m

        dz1 = np.dot(dz2, self.weights2.T) * self.a1 * (1 - self.a1)
        dw1 = np.dot(x.T, dz1) / m
        db1 = np.sum(dz1, axis=0) / m

        # Update weights and biases
        self.weights2 -= self.learning_rate * dw2
        self.bias2 -= self.learning_rate * db2
        self.weights1 -= self.learning_rate * dw1
        self.bias1 -= self.learning_rate * db1

    def train(self, x, y, epochs=1000):
        for epoch in range(epochs):
            output = self.forward(x)
            self.backward(x, y, output)
            if (epoch + 1) % 100 == 0:
                loss = -np.mean(np.log(output[range(y.shape[0]), y]))
                print(f"Epoch {epoch + 1}, Loss: {loss}")

    def predict(self, x):
        output = self.forward(x)
        return np.argmax(output, axis=1)


# Custom SVM implementation
class LinearSVM:
    def __init__(self, learning_rate=0.001, lambda_param=0.01, epochs=1000):
        self.learning_rate = learning_rate  # Learning rate
        self.lambda_param = lambda_param    # Regularization strength
        self.epochs = epochs               # Number of epochs
        self.weights = None                # Weight vector
        self.bias = None                   # Bias term

    def hinge_loss(self, x, y):
        """Compute hinge loss"""
        n_samples = x.shape[0]
        distances = 1 - y * (np.dot(x, self.weights) + self.bias)
        distances[distances < 0] = 0  # Hinge loss cannot be negative
        hinge_loss = self.lambda_param * (np.sum(distances) / n_samples)
        return hinge_loss

    def fit(self, x_train, y_train):
        """Train Linear SVM"""
        n_samples, n_features = x_train.shape
        self.weights = np.zeros(n_features)  # Initialize weights to zero
        self.bias = 0                        # Initialize bias to zero

        for epoch in range(self.epochs):
            for i in range(n_samples):
                # Hinge loss condition
                if y_train[i] * (np.dot(x_train[i], self.weights) + self.bias) < 1:
                    # Update weights and bias (for misclassified samples)
                    self.weights -= self.learning_rate * (self.lambda_param * -y_train[i] * x_train[i] + 2 * self.lambda_param * self.weights)
                    self.bias -= self.learning_rate * (self.lambda_param * -y_train[i])
                else:
                    # Only apply regularization for correctly classified samples
                    self.weights -= self.learning_rate * (2 * self.lambda_param * self.weights)

            # Optionally: print loss every 100 epochs
            if (epoch + 1) % 100 == 0:
                loss = self.hinge_loss(x_train, y_train)
                print(f"Epoch {epoch + 1}, Loss: {loss}")

    def predict(self, x):
        """Predict class labels"""
        linear_output = np.dot(x, self.weights) + self.bias
        return np.sign(linear_output)  # Return class labels (+1 or -1)


# Data preprocessing and loading
lfw_data = datasets.fetch_lfw_people(min_faces_per_person=70, resize=0.4)
features = lfw_data.data
labels = lfw_data.target

# Normalize data
scaler = StandardScaler()
features = scaler.fit_transform(features)

# Split the dataset
X_train, X_val, y_train, y_val = train_test_split(features, labels, test_size=0.2, random_state=42)

# Train the MLP classifier
mlp = MLPClassifier(input_dim=X_train.shape[1], hidden_dim=128, output_dim=len(np.unique(labels)), learning_rate=0.01)
mlp.train(X_train, y_train, epochs=2000)

# Evaluate MLP model
mlp_predictions = mlp.predict(X_val)
print("MLP Model Performance:")
accuracy = accuracy_score(y_val, mlp_predictions)
precision = precision_score(y_val, mlp_predictions, average='weighted')
recall = recall_score(y_val, mlp_predictions, average='weighted')
f1 = f1_score(y_val, mlp_predictions, average='weighted')
print(f"Accuracy: {accuracy}, Precision: {precision}, Recall: {recall}, F1 Score: {f1}")

# Train SVM classifier (convert labels for binary classification)
binary_labels = np.where(labels == 0, 1, -1)
X_train_bin, X_val_bin, y_train_bin, y_val_bin = train_test_split(features, binary_labels, test_size=0.2, random_state=42)

svm = LinearSVM(learning_rate=0.001, lambda_param=0.01, epochs=1000)
svm.fit(X_train_bin, y_train_bin)

# Evaluate SVM model
svm_predictions = svm.predict(X_val_bin)
print("SVM Model Performance (Custom Implementation):")
accuracy = accuracy_score(y_val_bin, svm_predictions)
precision = precision_score(y_val_bin, svm_predictions, average='weighted')
recall = recall_score(y_val_bin, svm_predictions, average='weighted')
f1 = f1_score(y_val_bin, svm_predictions, average='weighted')
print(f"Accuracy: {accuracy}, Precision: {precision}, Recall: {recall}, F1 Score: {f1}")
```
